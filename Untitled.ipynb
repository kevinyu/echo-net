```python
import numpy as np
import matplotlib.pyplot as plt
from echo.network import ReservoirState, Reservoir, ReservoirNetwork

%load_ext autoreload
%autoreload 2
```

```python
def pulses(N, pulse_size, n=100):
    y = np.zeros(N)
    for i in range(N//n):
        y[i*n:i*n+n] = np.random.choice([1, 2, 0, -1, -2])
    return y
```

```python
def pulse(total_duration, pulse_start, pulse_duration, dt=0.1):
    result = np.zeros(int(total_duration / dt))
    n = int(pulse_duration / dt)
    idx_start = int(pulse_start / dt)
    result[idx_start:idx_start + n] = 5
    return result
```

```python
res = Reservoir(800, N_in=2, N_out=1, sparsity=0.25, g=1.8, noise=0.01, feedback_on=1.0)
```

```python
plt.imshow(res.W_recurrent.toarray())
plt.clim(-1, 1)
plt.colorbar()
plt.title(r"$W_{recurrent}$")
plt.show()
```

```python
sys = ReservoirNetwork(res)
```

First lets drive it with no input (1000 timesteps)

```python
null_input_result = sys.run(input_data=np.zeros((2, 5000)))
```

```python
plt.plot(null_input_result["r"].T[:, :5])
plt.title("No input")
plt.show()
```

Next with a sine wave

```python
always = ReservoirState(np.random.normal(size=(800, 1)))
```

```python
t = np.linspace(0, 100, 20000)
sin = np.zeros((2, 20000))
sin[0, :] = np.sin(2 * np.pi * 2.0 * t)
sinehi_input_result = sys.run(input_data=sin, state=always)
```

```python
plt.plot(sinehi_input_result["r"].T[:, :5])
plt.title("High frequency input")
plt.show()
```
